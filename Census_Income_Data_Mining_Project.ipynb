5. Varying the Regularization Parameter in Logistic Regression
Plot: Accuracy vs. Regularization Parameter in Logistic Regression
 • Explanation: Regularization helps to prevent overfitting by penalizing large coefficients. The parameter C controls the strength of regularization. A very small C value (strong regularization) can lead to underfitting, while a very large C value (weak regularization) can lead to overfitting. An optimal C value balances these effects.


```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
C_values = [0.01, 0.1, 1, 10, 100]
accuracies = []

for C in C_values:
    lr = LogisticRegression(C=C, random_state=42)
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

plt.figure(figsize=(10, 6))
plt.plot(C_values, accuracies, marker='o')
plt.title('Accuracy vs. Regularization Parameter in Logistic Regression')
plt.xlabel('C (Regularization Parameter)')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid(True)
plt.show()
```

```python
model = LogisticRegression(C=100)

# Training
model.fit(X_train, y_train)

# Predictions
y_pred1 = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred1)
print("Accuracy:", accuracy)

mse = mean_squared_error(y_test, y_pred1)
print("Mean Squared Error:", mse)

# Calculate Root Mean Squared Error
rmse = sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculate R^2 Score
r2 = r2_score(y_test, y_pred1)
print("R2 Score:", r2)

precision = precision_score(y_test, y_pred1, average='macro')  # 'macro' is one method of handling multi-class
print(f'Precision : {precision}')

recall = recall_score(y_test, y_pred1, average='macro')
print(f'Recall Score : {recall}')

f1 = f1_score(y_test, y_pred1, average='macro')
print(f'F1 Score : {f1}')

conf_matrix = confusion_matrix(y_test, y_pred1)
print("Confusion Matrix:\n", conf_matrix)

disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
